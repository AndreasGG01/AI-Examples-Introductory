```python
start_point = 'Goodsprings'
```

```python
def fitness_fn_for_TSP(sample):
    """ total distance for the Traveling Salesman to be covered if in sample  """
    cost = 0
    for i in range(len(sample) - 1):
        cost += distances[sample[i]][sample[i + 1]]
    cost += distances[sample[0]][sample[-1]]
    return cost
```

```python
gene_pool = all_cities
max_population = 200
mutation_rate = 0.5
number_of_elements_in_gene_pool = 19
```
